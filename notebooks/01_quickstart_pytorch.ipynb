```python
import numpy as np
from netforge_rl.environment.parallel_env import NetForgeRLEnv

env = NetForgeRLEnv({'scenario_type': 'ransomware', 'max_ticks': 50})
obs, _ = env.reset(seed=0)
rng = np.random.default_rng(0)

totals = {a: 0.0 for a in env.possible_agents}
while env.agents:
    actions = {a: np.array([rng.integers(0, n) for n in env.action_space(a).nvec], dtype=np.int64) for a in env.agents}
    obs, rewards, term, trunc, _ = env.step(actions)
    for a, r in rewards.items():
        totals[a] += r
    if all(term.values()) or all(trunc.values()):
        break

totals
```
