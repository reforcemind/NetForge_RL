# 02 — Visualize an attack as a GIF

Drive the env with a scripted Red agent and the renderer's `FrameRecorder`.
Output: `attack.gif` showing node colors flipping green → red → grey across the episode.

```python
import numpy as np
from netforge_rl.baselines import HeuristicRedPolicy, RandomPolicy
from netforge_rl.environment.parallel_env import NetForgeRLEnv
from netforge_rl.render import FrameRecorder

env = NetForgeRLEnv({'scenario_type': 'ransomware', 'max_ticks': 60})
env.reset(seed=42)
red = HeuristicRedPolicy(seed=0)
blue = RandomPolicy(seed=1)
rec = FrameRecorder(fps=4)

for _ in range(60):
    if not env.agents:
        break
    rec.append(env.render(mode='rgb_array'))
    actions = {}
    for agent in env.agents:
        actions[agent] = (red if 'red' in agent else blue).act(env, agent)
    env.step(actions)

rec.save('attack.gif')
```
