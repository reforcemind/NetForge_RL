# 04 — JAX PPO at 4096 parallel envs

Headline throughput demo. On a CPU we already see ≥10⁶ aggregate SPS; on an A100 expect another order of magnitude. Trains the hand-rolled PureJaxRL-style PPO in `netforge_rl.baselines.jax_ppo`.

```python
import jax
print('JAX devices:', jax.devices())

from netforge_rl.baselines.jax_ppo import PPOConfig, train
out = train(PPOConfig(total_iters=10, num_steps=32, batch_size=4096, n_hosts=100))
out['losses']
```

```python
import matplotlib.pyplot as plt
plt.plot(out['losses']); plt.xlabel('iter'); plt.ylabel('loss'); plt.title('PPO @ 4096 envs')
```
